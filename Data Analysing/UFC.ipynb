# Splitting significant strike attempted and landed in separate columns.

```python
ufc['R_SIG_STR_ATTEMPT'] = ufc['R_SIG_STR.'].apply(lambda x : int(x.split('of')[-1]))
```

```python
ufc['R_SIG_STR_LANDED'] = ufc['R_SIG_STR.'].apply(lambda x : int(x.split('of')[0]))
```

```python
ufc['B_SIG_STR_ATTEMPT'] = ufc['B_SIG_STR.'].apply(lambda x : int(x.split('of')[-1]))
```

```python
ufc['B_SIG_STR_LANDED'] = ufc['B_SIG_STR.'].apply(lambda x : int(x.split('of')[0]))
```

# Deleting R_SIG_STR AND B_STG_STR columns  

```python
ufc.drop(['R_SIG_STR.', 'B_SIG_STR.'], axis=1, inplace=True)
```

# Same for TOTAL STRIKES BY BLUE AND RED TEAM 

```python
ufc['R_TOTAL_STR_ATTEMPT'] = ufc['R_TOTAL_STR.'].apply(lambda x : int(x.split('of')[-1]))
ufc['R_TOTAL_STR_LANDED'] = ufc['R_TOTAL_STR.'].apply(lambda x : int(x.split('of')[0]))
```

```python
ufc['B_TOTAL_STR_ATTEMPT'] = ufc['B_TOTAL_STR.'].apply(lambda x : int(x.split('of')[-1]))
ufc['B_TOTAL_STR_LANDED'] = ufc['B_TOTAL_STR.'].apply(lambda x : int(x.split('of')[0]))
```

```python
ufc.drop(['R_TOTAL_STR.', 'B_TOTAL_STR.'], axis=1, inplace=True)
```

# Red team vs Blue team wins 

```python
ufc['Winner'].isnull().sum()  # if Null values
```

```python
def red_blue_draw(row):
    
    if row['R_fighter'] == row['Winner']:
        return 'red'
    elif row['B_fighter'] == row['Winner']:
        return 'blue'
    else:
        return 'draw'
```

```python
ufc['team'] = ufc[['R_fighter', 'B_fighter', 'Winner']].apply(red_blue_draw, axis=1)
```

```python
ufc['team'].isnull().sum()
```

```python
ufc.Winner.fillna('draw', inplace=True)
```

```python
ufc.Winner.notnull().value_counts()
```

```python
plt.pie(ufc.team.value_counts(), explode=[0.05,0.05,0.05], autopct="%1.1f%%", labels=['red', 'blue', 'draw'], colors=['orange', 'skyblue', 'cyan'])
```

# Win By  

```python
# Win types
ufc.groupby(ufc.win_by).size()
```

```python
reds_win_by = ufc.win_by[ufc['team'] == 'red'].value_counts().sort_index()
```

```python
blues_win_by = ufc.win_by[ufc['team'] == 'blue'].value_counts().sort_index()
```

```python
draws_by = ufc.win_by[ufc['team'] == 'draw'].value_counts().sort_index()
```

```python
plt.subplots_adjust(bottom=10, top=20)
fig = plt.figure(dpi=100, figsize=(14, 16))
rect = fig.patch
rect.set_facecolor('silver')
graph1 = fig.add_subplot(3,1,1, fc='black', title='Red Wins By')
graph1.bar(reds_win_by.index, reds_win_by)
graph2 = fig.add_subplot(3,1,2, fc='black', title='Blue Wins By')
graph2.bar(blues_win_by.index, blues_win_by)
graph3 = fig.add_subplot(3,1,3, fc='black', title='Draws By')
graph3.bar(draws_by.index, draws_by) 
```
