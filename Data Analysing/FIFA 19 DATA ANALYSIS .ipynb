Why Majority Players are rated 63 - 68 ? 

```python
less_rated = fifa_df[(fifa_df['Overall'] >= 63) & (fifa_df['Overall'] <= 68)]
```

```python
less_rated.shape
```

```python
less_rated.head(10)
```

# From where they belong

```python
fifa_df.Nationality.isnull().sum()
```

```python
less_rated.groupby('Nationality').size()
```

### TOP 10 less rated players by country

```python
plt.figure(dpi=100)
less_rated.groupby('Nationality').size().sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%', explode=np.arange(10)*0.02)
plt.title('Less rated Players by Countries', pad=14)
plt.ylabel('')
plt.legend(title='Countries', loc='upper right', bbox_to_anchor=(0.10, 0.10))
```

```python
less_rated.groupby('Age').size().plot()
```
